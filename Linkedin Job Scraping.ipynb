```python
from webdriver_manager.chrome import ChromeDriverManager #with the help of chromedriver_manager, it will automatically install chromedriver as per your google version
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
#driver=webdriver.Chrome(executable_path='C:\Program Files\Google\Chrome\Application\chrome.exe')
#driver=webdriver.Chrome()
```

```python
url='https://www.linkedin.com/jobs/search/?currentJobId=3778767989&keywords=data%20scientist&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true&sortBy=R'
driver.get(url)
driver.implicitly_wait(5)#it will take 5s to load the data of the url
```

```python
n=driver.find_elements_by_class_name('results-context-header__job-count')[0].text
n
```

```python
i=2
while i<=16:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #this helps to scroll down from top to bottom
    i=i+1

    try:
        x=driver.find_element_by_xpath("//button[@aria-label='See more jobs']") #this helps to click on the see more jobs button
        driver.execute_script("arguments[0].click();",x)
        time.sleep(5)
    except:
        pass
        time.sleep(5)
```

```python
n=n.replace('+','')
n=n.replace(',','')
n
```

```python
y=pd.to_numeric(n)
y
```

```python
companyname=[]
```

```python
for j in range(200):
    try:
        company=driver.find_elements_by_class_name('base-search-card__subtitle')[j].text
        companyname.append(company)
        
    except IndexError:
        print("Done")
```

```python
companyname
```

```python
titlename=[]
```

```python
for j in range(200):
    try:
        title=driver.find_elements_by_class_name('base-search-card__title')[j].text
        titlename.append(title)
        
    except IndexError:
        print("Done")
```

```python
titlename
```

```python
companyfinal = pd.DataFrame(companyname,columns=['Company'])
titlefinal=pd.DataFrame(titlename,columns=['Title'])
```

```python
final=companyfinal.join(titlefinal)
```

```python
final
```

```python
linklist=[]
```

```python
findlink=driver.find_elements_by_class_name('base-card__full-link')
for k in findlink:
    linklist.append(k.get_attribute('href'))
```

```python
linklist
```

```python
final_linklist = pd.DataFrame(linklist, columns=['link'])
```

```python
final_linklist=final_linklist.head(200)
```

```python
final = final.join(final_linklist)
```

```python
final
```

```python
datelist=[]
```

```python
for i in range(200):
    try:
        date=driver.find_elements_by_xpath("//time[@class='job-search-card__listdate']")[i].text
        datelist.append(date)
    except:
        datelist.append('NA')
```

```python
datelist
```

```python
final_date= pd.DataFrame(datelist,columns=['Requirement since'])
```

```python
final_date
```

```python
final=final.join(final_date)
```

```python
final
```
